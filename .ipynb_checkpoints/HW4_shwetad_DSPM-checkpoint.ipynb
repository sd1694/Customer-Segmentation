# Customer Segmentation using k-means clustering

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Step 1 - Data Pre-processing

```python
#load all three datasets
customers = pd.read_excel(open('bikeshops.xlsx','rb'), sheet_name='Sheet1')
products = pd.read_excel(open('bikes.xlsx','rb'), sheet_name='Sheet1')
orders = pd.read_excel(open('orders.xlsx','rb'), sheet_name='Sheet1')
```

```python
#merged datasets
orders_total = pd.merge(orders, customers, left_on = "customer.id", right_on="bikeshop.id")
orders_total = pd.merge(orders_total, products, left_on = "product.id", right_on = "bike.id")
orders_total.head()
```

```python
#to free memory space
del customers
del products
del orders
```

```python
orders_total['price.extended'] = orders_total['price'] * orders_total['quantity'] 
column_names = ['order.date', 'order.id', 'order.line', 'bikeshop.name', 'model', 'quantity', 'price', 'price.extended', 
                'category1', 'category2', 'frame']
orders_total_final = orders_total[column_names]
orders_total_final = orders_total_final.sort_values(['order.id','order.line'], ascending=[True, True])
orders_total_final.head()
```

```python
#converted the unit price to categorical high/low variables; 1-high price, 0-low price
median_price = orders_total_final['price'].median()
orders_total_final['price'] = np.where((orders_total_final['price']>=median_price), 1, 0)
orders_total_final.head()
```

```python
#Customer spread by quantity of bike models purchased
customer_trends = orders_total_final.groupby(['bikeshop.name', 'model', 'category1',
                                              'category2', 'frame', 'price'])['quantity'].sum().reset_index()
pivot_data = pd.pivot_table(customer_trends, index=['model', 'category1', 'category2', 'frame', 'price'], 
                            columns=['bikeshop.name'], aggfunc=sum, fill_value=0)
pivot_data.head()
```

```python
#scaling of the data
total_bikes = pivot_data.sum()
pivot_data_final = pivot_data/total_bikes
pivot_data_final.head()
```

## Step 2 - Perform KMeans Clustering

```python
kmeans_data = pivot_data_final.reset_index(drop=True).T
kmeans_data
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

best_k = None
best_labels = None
best_score = -1
best_model = None
best_cluster = None
silhouette_score_list =[]
min_k = 4
max_k = 8

for k in range(min_k, max_k+1):
    kmeans = KMeans(n_clusters=k, n_init=100, random_state=95865)
    kmeans.fit(kmeans_data)
    cluster_assignments = kmeans.predict(kmeans_data)
    labels = kmeans.labels_
    score = silhouette_score(kmeans_data, labels)
    silhouette_score_list.append(score)
    print('k =', k, 'has silhouette score:', score)
    
    if score > best_score:
        best_score = score
        best_k = k
        best_model = kmeans
        best_labels = labels
        best_cluster = cluster_assignments
    
print()
print("Best k = " + str(best_k), 'as per silhouette score')
print()
```

```python
plt.style.use('seaborn-darkgrid')
plt.plot(range(min_k,max_k+1), silhouette_score_list,marker='o')
plt.title('Silhouette Score Plot', fontsize = 10)
plt.xlabel('k',fontsize = 10)
plt.ylabel('Silhouette Score', fontsize = 10)
```

From our hypothesis, we expect there to be at least four groups of customers. This is because customer preference is expected to vary by price (high/low) and category1 (mountain vs bike). We also believe there could be more as some customers may not care about price but may still prefer a specific bike category. Beyond eight segments may be overfitting the segments.

The silhouette value is a measure of how similar an object is to its own cluster (cohesion) compared to other clusters (separation). The silhouette ranges from âˆ’1 to +1, where a high value indicates that the object is well matched to its own cluster and poorly matched to neighboring clusters. Hence, best k is the one with the maximum silhouette score which is 5 in our case.

## Step 3 - Analyze the result

```python
#names in each segment
bikeshop_names = kmeans_data.T['quantity'].columns
for i in range(0,best_k):
    print('[ Bikeshops in Cluster:', i,']')
    for j in range(0,len(best_labels)):
        if i==best_labels[j]:
            print(bikeshop_names[j])
    print()
```

```python
centroids = pd.DataFrame(best_model.cluster_centers_.T)
df_index = (pivot_data_final.index).to_frame()
df_index = df_index.reset_index(drop=True)
df_merged = pd.concat([df_index, centroids], axis=1)
df_merged.head()
```

```python
#preferences of segments
from IPython.display import display, HTML

for i in range(0,best_k):
    df_merged = df_merged.rename(columns = {i: ("Cluster"+str(i))})

    
for i in range(0,best_k):
    print('Top 10 bike models for Cluster:',i)
    df_sorted = df_merged.sort_values(by=["Cluster"+str(i)], ascending=False)
    columns_to_drop = ["Cluster"+str(j) for j in range(0,best_k) if j!=i]
    df_sorted = df_sorted.drop(columns_to_drop,axis=1)
    display(df_sorted.head(10))
```

Interpretation of bike models in cluster:

1. **Cluster 0:** We can see that the bike models in this cluster majorly have aluminium frame and are low-end models.
2. **Cluster 1:** This cluster tends to prefer low-end road bikes with a mix of frames.
3. **Cluster 2:** These are all top-end road models. In addition, all but one model has a carbon frame.
4. **Cluster 3:** Top models in this cluster are all low-end/affordable models. Thereâ€™s a mix of road and mountain in category1 and a mix of frame material as well.
5. **Cluster 4:** We can see that the top 10 models purchased are predominantly high-end and mountain. All but one model has a carbon frame.

We got clear trends for four of five groups, but two groups (clusters 1 and 3) are very similar. Because of this, it may make sense to combine these two groups or to switch from k = 5 to k = 4 results.

### Gausian Mixture Model - GMM

```python
from sklearn.mixture import GaussianMixture
from sklearn.metrics import calinski_harabaz_score

best_score_g = -np.inf
best_k_g = None
best_model_g = None
best_cluster_assignments_g = None
min_k = 4
max_k = 8

for k in range(min_k, max_k+1):
    candidate_gmm = GaussianMixture(n_components=k, n_init=100, random_state=0)
    candidate_gmm.fit(kmeans_data)
    cluster_assignments = candidate_gmm.predict(kmeans_data)
    score = calinski_harabaz_score(kmeans_data, cluster_assignments)
    
    print('k =', k, 'has CH index', score)
    
    if score > best_score_g:
        best_score_g = score
        best_k_g = k
        best_model_g = candidate_gmm
        best_cluster_assignments_g = cluster_assignments
    
print('Best k within 2, 3, ..., 10 according to CH index:', best_k_g)
```

```python
h_labels = best_cluster_assignments_g

for i in range(0,best_k_g):
    print('[ Bikeshops in Cluster:', i,']')
    for j in range(0,len(h_labels)):
        if i==h_labels[j]:
            print(bikeshop_names[j])
    print()
```

Using GMM, the best CH index is 4 i.e. 4 clusters would best represent this data. Cluster 2 and 3 of GMM are exactly the same as Cluster 2 and 4 of KMeans. Since the number of clusters is one less here, Philadelphia Bike Shop and San Antonio Bike Shop of Cluster 3 have moved to other clusters here. Philadelphia Bike Shop is present in Cluster 0 of GMM and San Antonio Bike Shop in Cluster 1. Additionally, New York Cycles has moved from Cluster 0 to 1 in GMM.

### Hierarchial Clustering - Divisive

```python
from sklearn.cluster import KMeans
from sklearn.metrics import calinski_harabaz_score

cluster_assignment = np.zeros(len(kmeans_data)).astype(np.int) 
data = kmeans_data

best_score_h = -np.inf
best_k_h = None
best_model_h = None
best_labels_h = None
best_cluster_assignments_h = None
num_splits = 8  

for split_idx in range(num_splits):
    num_clusters = cluster_assignment.max() + 1
    max_cluster_RSS = -np.inf
    arg_max = None
    for cluster_idx in range(num_clusters):
        cluster_mask = (cluster_assignment == cluster_idx)
        cluster_size = cluster_mask.sum()
        if cluster_size == 1:
            cluster_RSS = 0.
        else:
            points_in_cluster = data[cluster_mask]
            cluster_mean = points_in_cluster.mean(axis=0)
            cluster_RSS = 0.
            for point in points_in_cluster:
                cluster_RSS += np.linalg.norm(point - cluster_mean)**2
                
        if cluster_RSS > max_cluster_RSS:
            max_cluster_RSS = cluster_RSS
            arg_max = cluster_idx
            
    kmeans = KMeans(n_clusters=2, n_init=1000, random_state=0)
    split_assignments = kmeans.fit_predict(data[cluster_assignment == arg_max])

    cluster_assignment[cluster_assignment == arg_max] \
        = arg_max * (split_assignments == 0) + num_clusters * (split_assignments == 1)
    
    score = calinski_harabaz_score(data, cluster_assignment)
    
    if cluster_assignment.max()>=3:
        print('k =', cluster_assignment.max()+1, 'has CH score:', score)
        if score > best_score_h:
            best_score_h = score
            best_k_h = cluster_assignment.max()+1
            best_model_h = kmeans
            best_labels_h = best_model_h.labels_
            best_cluster_assignments_h = cluster_assignment.copy()
```

Since k=4 has the maximum CH score, we have 4 clusters.

```python
h_labels = best_cluster_assignments_h

for i in range(0,best_k_h):
    print('[ Bikeshops in Cluster:', i,']')
    for j in range(0,len(h_labels)):
        if i==h_labels[j]:
            print(bikeshop_names[j])
    print()
```

Here, a lot of bikeshops seemed to have combined in Cluster 0. Cluster 1 and 2 are pretty similar to the ones we got in KMeans.

```python
agglo_labels = cluster.labels_

for i in range(0,number_of_clusters):
    print('[ Bikeshops in Cluster:', i,']')
    for j in range(0,len(agglo_labels)):
        if i==agglo_labels[j]:
            print(bikeshop_names[j])
    print()
```

If I consider 5 clusters in Agglomerative clustering, I get the exact same results as I got in KMeans.

```python
#Agglomerative with 4 clusters
from sklearn.cluster import AgglomerativeClustering

number_of_clusters = 4
cluster = AgglomerativeClustering(n_clusters=number_of_clusters, affinity='euclidean', linkage='ward')  
x = cluster.fit_predict(kmeans_data)
```

```python
agglo_labels = cluster.labels_

for i in range(0,number_of_clusters):
    print('[ Bikeshops in Cluster:', i,']')
    for j in range(0,len(agglo_labels)):
        if i==agglo_labels[j]:
            print(bikeshop_names[j])
    print()
```

If we consider 4 clusters in Agglomerative Clustering, the results are exactly the same except Cluster 3 of KMeans combines with Cluster 0 here and is represented as one cluster here (Cluster 0).
